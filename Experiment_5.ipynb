```python
# Navya Bhargava - 25070123079
# Study of Dictionaries in Python
# Dictionaries are used to store data values in key:value pairs.
#A dictionary is a collection which is ordered*, changeable and does not allow duplicates.
```

```python
# Creation of Dictionary
thisdict={"brand":"Ford",
          "model":"Mustang",
          "Year":1964
          }
print(thisdict)
```

```python
# Retriving values from dictionary
thisdict={"brand":"Ford",
          "model":"Mustang",
          "Year":1964
          }
print(thisdict["brand"])
```

```python
# Dictionary takes latest value given for key. It will not allow diulpicate
thisdict={"brand":"Ford",
          "model":"Mustang",
          "Year":1964,
          "Year":2020
          }
print(thisdict)
```

```python
# Length of key-value pairs
len(thisdict)
```

```python
# Changing/Adding Key:value in dictionary
car={"brand":"Ford",
          "model":"Mustang",
          "Year":1964,
          "Year":2020
          }
car["color"]="red"
print(car)
x = car.values()
print(x)
print(type(car))
```

```python
# Removing Elements
student = {
    "roll":80,
    "name":"NEIL",
    "branch":"E&TC"
}
student.pop("branch")
print(student)
```

```python
#1. A dictionary stores product names and their prices.
# Write a python program to update the price of a given product.
products = {"Pen": 10, "Book": 50, "Scale":20}
print("Original Prodcts:", products)
products["Book"] = 55
print("Updated Products:", products)
```

```python
# 2. A dictionary stores student names and marks.
# Write a Python program to search for a student's mark.
# If the student is not found, display a suitable message.
marks = {"A": 85, "B": 72, "C": 90}
name=input("Enter the name of the student: ")
print(marks.get(name,"Student not found"))
```

```python
#3. A dictionary stores username and passwords.
#Write a Python program to validate user login.
users = {"user1": "password1", "user2": "password2", "user3": "password3"}
u=input("Enter the username: ")
if u in users:
    p=input("Enter the password: ")
    if users[u]==p:
     print("Login Successful")
    else:
     print("Invalid Credentials")
else:
    print("User not found")
```

```python
#4. A dictionary stores students names and marks.
#Write a Python to find the student who scored the higest marks.
marks = {"A": 85, "B": 92, "C": 78}
topper=max(marks,key=marks.get)
print("Topper:",topper)
print("Marks:",marks[topper])
```
